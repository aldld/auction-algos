```python
import auction
import math

canvas_width = 800
canvas_height = 500

tree_height = 4
num_nodes = 2**tree_height - 1

nodes = [auction.Node("0", canvas_width // 2, 50, 1.0)]

for i in range(1, num_nodes):
    level = math.frexp(i + 1)[1] - 1
    offset = (canvas_width - 50) / (2**(level+1))
    parent = (i - 1) // 2
    
    x_pos = nodes[parent].x_pos + (offset if i % 2 == 1 else -offset)
    y_pos = nodes[parent].y_pos + 100
    
    surplus = 0.0
    if level == tree_height - 1:
        surplus = -1 / (2**(tree_height - 1))
    
    nodes.append(auction.Node(str(i), x_pos, y_pos, surplus))
    
graph = auction.Graph(nodes)

for i in range(1, num_nodes):
    parent = (i - 1) // 2
    graph.add_edge(nodes[parent], nodes[i], 1.0)
    graph.add_edge(nodes[i], nodes[parent], 1.0)

auction.er_min_cost_flow(
    eps=0.25,
    graph=graph,
    debug=True,
    visualizer=auction.IPyDisplayD3(eps=0.25, width=canvas_width, height=canvas_height))
```
